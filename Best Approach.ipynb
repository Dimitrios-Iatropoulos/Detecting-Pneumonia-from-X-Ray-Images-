```python
import numpy as np
import pandas as pd
from tensorflow.keras import layers, models, Input, callbacks, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.utils.class_weight import compute_class_weight
from sklearn.model_selection import KFold
import matplotlib.pyplot as plt
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.layers import GlobalAveragePooling2D, Flatten, BatchNormalization, Dense, Dropout
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam

# Load CSV file with image names and labels
train_csv_path = "/content/labels_train.csv"
train_df = pd.read_csv(train_csv_path)

# Define image directory
train_image_dir = "/content/train_images/train_images"

# Convert label to string for use in generators
train_df['class_id'] = train_df['class_id'].astype(str)

num_folds = 5
kfold = KFold(n_splits=num_folds, shuffle=True, random_state=42)

# Data augmentation for training
data_augmentation = ImageDataGenerator(
    rescale=1./255,
    rotation_range=7,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    fill_mode="nearest"
)

val_datagen = ImageDataGenerator(rescale=1./255)

# Compute class weights
class_labels = np.unique(train_df['class_id'].astype(int))
class_weights = compute_class_weight('balanced', classes=class_labels, y=train_df['class_id'].astype(int))
class_weight_dict = dict(zip(class_labels, class_weights))

# Model architecture
input_shape = (224, 224, 3)
input_layer = Input(shape=input_shape)
densenet_base = DenseNet121(weights='imagenet', include_top=False, input_tensor=input_layer)

# Unfreeze the last 30 layers
for layer in densenet_base.layers[-30:]:
    layer.trainable = True

# Function to build the model
def build_model():
    model_densenet = GlobalAveragePooling2D()(densenet_base.output)
    output_densenet = Flatten()(model_densenet)

    x = BatchNormalization()(output_densenet)
    x = Dense(256, activation='relu', kernel_regularizer=l2(0.001))(x)
    x = Dropout(0.5)(x)
    x = BatchNormalization()(x)
    x = Dense(128, activation='relu', kernel_regularizer=l2(0.001))(x)
    x = Dropout(0.5)(x)
    x = BatchNormalization()(x)
    x = Dense(64, activation='relu', kernel_regularizer=l2(0.001))(x)
    x = Dropout(0.5)(x)
    x = BatchNormalization()(x)
    output_layer = Dense(3, activation='softmax')(x)

    model = models.Model(inputs=input_layer, outputs=output_layer)
    model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Print the model summary
model = build_model()
model.summary()

# Hyperparameters used
hyperparameters = {
    "learning_rate": 0.0001,
    "batch_size": 32,
    "epochs": 25,
    "optimizer": "Adam",
    "loss_function": "categorical_crossentropy",
    "early_stopping_patience": 7,
    "reduce_lr_factor": 0.2,
    "reduce_lr_patience": 5,
    "class_weight": class_weight_dict
}

print("Hyperparameters used:")
for key, value in hyperparameters.items():
    print(f"{key}: {value}")

# Setup callbacks
early_stopping = callbacks.EarlyStopping(monitor='val_accuracy', patience=7, verbose=1, restore_best_weights=True)
reduce_lr = callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, verbose=1, min_delta=0.0001)

# Function to get checkpoint path
def get_checkpoint_path(fold):
    return f"/content/drive/MyDrive/trial/fold-{fold}-epoch-{{epoch:04d}}-cp.ckpt"

# Lists to hold accuracy and loss for all folds
train_accuracies = []
val_accuracies = []
train_losses = []
val_losses = []

# Iterate through K folds
for fold, (train_index, val_index) in enumerate(kfold.split(train_df)):
    train_fold_df = train_df.iloc[train_index]
    val_fold_df = train_df.iloc[val_index]

    # Create generators for training and validation
    train_generator = data_augmentation.flow_from_dataframe(
        dataframe=train_fold_df,
        directory=train_image_dir,
        x_col="file_name",
        y_col="class_id",
        target_size=(224, 224),
        batch_size=32,
        class_mode="categorical",
        shuffle=True,
        seed=42
    )

    val_generator = val_datagen.flow_from_dataframe(
        dataframe=val_fold_df,
        directory=train_image_dir,
        x_col="file_name",
        y_col="class_id",
        target_size=(224, 224),
        batch_size=32,
        class_mode="categorical",
        seed=42
    )

    # Weight Saving
    cp_callback = callbacks.ModelCheckpoint(
      filepath=get_checkpoint_path(fold),
      save_weights_only=True,
      save_best_only=True,
      monitor='val_loss',
      verbose=1
    )

    # train model
    history = model.fit(
        train_generator,
        steps_per_epoch=len(train_fold_df) // 32,
        epochs=25,
        validation_data=val_generator,
        validation_steps=len(val_fold_df) // 32,
        class_weight=class_weight_dict,
        callbacks=[early_stopping, cp_callback, reduce_lr]
    )

    # Append accuracies and losses
    train_accuracies.append(history.history['accuracy'])
    val_accuracies.append(history.history['val_accuracy'])
    train_losses.append(history.history['loss'])
    val_losses.append(history.history['val_loss'])

    # Plot accuracy scores for current fold
    plt.figure(figsize=(12, 6))
    plt.plot(train_accuracies[-1], label=f'Fold {fold+1} Train Accuracy')
    plt.plot(val_accuracies[-1], label=f'Fold {fold+1} Val Accuracy')
    plt.title(f'Train and Validation Accuracy for Fold {fold+1}')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend(loc='upper left')
    plt.show()

    # Plot loss scores for current fold
    plt.figure(figsize=(12, 6))
    plt.plot(train_losses[-1], label=f'Fold {fold+1} Train Loss')
    plt.plot(val_losses[-1], label=f'Fold {fold+1} Val Loss')
    plt.title(f'Train and Validation Loss for Fold {fold+1}')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend(loc='upper left')
    plt.show()
```

```python
import numpy as np
import pandas as pd
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from google.colab import drive


# Define the paths to the best weights
weights_paths = [
    "/content/drive/MyDrive/trial/fold-4-epoch-0012-cp.ckpt",
    "/content/drive/MyDrive/trial/fold-4-epoch-0011-cp.ckpt",
    "/content/drive/MyDrive/trial/fold-3-epoch-0014-cp.ckpt",
    "/content/drive/MyDrive/trial/fold-3-epoch-0007-cp.ckpt",
    "/content/drive/MyDrive/trial/fold-1-epoch-0005-cp.ckpt",
    "/content/drive/MyDrive/trial/fold-4-epoch-0005-cp.ckpt",
    "/content/drive/MyDrive/trial/fold-4-epoch-0002-cp.ckpt",
    "/content/drive/MyDrive/trial/fold-3-epoch-0010-cp.ckpt",
    "/content/drive/MyDrive/trial/fold-2-epoch-0006-cp.ckpt",
    "/content/drive/MyDrive/trial/fold-1-epoch-0001-cp.ckpt",

]

# Make predictions with each model
all_predictions = []

# image size and batch size
img_width, img_height = 224, 224
batch_size = 32

test_datagen = ImageDataGenerator(rescale=1./255)

test_dir = "/content/test_images"
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode=None,
    shuffle=False
)

# Loop through each set of weights and make predictions
for weights_path in weights_paths:
    # Load the weights into the pre-initialized model
    model.load_weights(weights_path)

    # Make predictions
    predictions = model.predict(test_generator)
    all_predictions.append(predictions)

# Convert list of predictions to NumPy array
all_predictions = np.array(all_predictions)

# Average the predictions across the models
average_predictions = np.mean(all_predictions, axis=0)

# Get the predicted class indices
y_pred = np.argmax(average_predictions, axis=1)

# Get filenames from the generator
file_names = test_generator.filenames

# Extract image names from filenames
image_names = [os.path.basename(file_name) for file_name in file_names]

# Create a DataFrame with image names and predicted labels
results_df = pd.DataFrame({"file_name": image_names, "class_id": y_pred})

# Save predictions to CSV
output_path = "/content/drive/MyDrive/predictions3.csv"
results_df.to_csv(output_path, index=False)
```
